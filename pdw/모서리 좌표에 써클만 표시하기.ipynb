```python
import numpy as np
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
img = cv2.imread('10-5.png')
```

```python
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
```

```python
ret,thresh = cv2.threshold(imgray,127,255,0) 
```

```python
im2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) 
## cv2.CHAIN_APPROX_SIMPLE 모서리의 4개점을 가지고 있는 좌표 값 
```

```python
print contours 
## vstack을 하기전 contours는 일반적으로 array 구조를 가진다 
```

```python
contours = np.vstack(contours).squeeze()
##cv2.findContours는 실제로 2 개의 항목을 반환합니다. 여기 "윤곽선"이 목록입니다. 
##따라서 우리는 numpy의 vstack() 을 사용하여 이들을 서로 쌓은 다음 squeeze() 를 사용하여 중복 축을 제거합니다.
```

```python
print contours
```

```python
for location in contours:
    x,y=location.ravel()
    cv2.circle(img, (x,y), 3, (0,255,0)) 
```
