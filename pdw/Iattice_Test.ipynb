```python
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
iatticeImage = cv2.imread('Iattice.jpg')
plt.imshow(iatticeImage)
```

```python
plt.imsave('Iattice_Test.jpg',iatticeImage)
```

```python
iatticeGrayImage = cv2.cvtColor(iatticeImage, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(iatticeGrayImage, cmap = "gray")
```

```python
fixedLength = 24
# 24pixel = ?cm
```

```python
height, width = iatticeGrayImage.shape
print height , width
```

```python
for x in range(0,width) :
    for y in range(0,height) :
        if iatticeGrayImage[x][y]>250 :
            iatticeGrayImage[x][y]=255
        elif iatticeGrayImage[x][y]<=5 :
            iatticeGrayImage[x][y]=0
```

```python
iatticeDrawingImage = cv2.cvtColor(iatticeImage, cv2.COLOR_RGB2GRAY)
plt.imshow(iatticeDrawingImage)
```

```python
count = 0
countyCheck = [ 0 for total in range(500)]
iatticePointer = [[0 for col in range(3)] for row in range(1000)]
countx=0
for x in range(1,width-1) :
    county=0
    for y in range(1,height-1) :
        #iatticeDrawingImage[x][y] = iatticeGrayImage[x][y]
        if iatticeGrayImage[x][y] == 255 and iatticeGrayImage[x-1][y-1] == 0 and iatticeGrayImage[x-1][y] == 255 and iatticeGrayImage[x][y-1] == 255 :
            count = count + 1
            iatticePointer[count][0] = x
            iatticePointer[count][1] = y
            iatticePointer[count][2] = countx
            county = county + 1
            for tempx in range(x-1,x+2) :
                for tempy in range(y-1,y+2) :
                    iatticeDrawingImage[tempx][tempy]=150
    if county != 0 :
        countx = countx + 1
        countyCheck[countx] = county
```
