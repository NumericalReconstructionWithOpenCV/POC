```python
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
cornerImage = plt.imread("10-01.png")
```

```python
plt.imshow(cornerImage)
```

```python
cornerGrayImage = cv2.cvtColor(cornerImage, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cornerGrayImage, cmap = "gray")
```

```python
cornerGrayImageArray = np.float32(cornerGrayImage)
print cornerGrayImageArray
```

```python
detectedCornerResult = cv2.cornerHarris(cornerGrayImageArray, 20, 3, 0.04)
print detectedCornerResult
```

```python
detectedCornerResult = cv2.dilate(detectedCornerResult, None)
print detectedCornerResult
```

```python
cornerImage[detectedCornerResult > 0.01 * detectedCornerResult.max()] = [0, 0, 255, 0]
```

```python
plt.imshow(cornerGrayImage, cmap = "gray")
```
