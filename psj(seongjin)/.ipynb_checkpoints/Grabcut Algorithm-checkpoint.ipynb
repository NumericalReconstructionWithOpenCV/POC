```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('messi.jpg')
mask = np.zeros(img.shape[:2],np.uint8)

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (60,109,350,700)

#지금까지 cv2, numpy 및 matplotlib를 가져 왔습니다. 
#그런 다음 이미지에로드하고, 마스크를 만들고, 알고리즘에 의해 내부적으로 사용되는 배경 및 전경 모델을 지정합니다. 
#중요한 부분은 rect입니다. rect = (start_x, start_y, width, height)입니다.
#rect는 우리의 주요 객체를 둘러싼 사각형입니다. 이미지를 사용하고 있다면 바로 사용할 수 있습니다. 
#당신의 이미지를위한 적절한 좌표를 찾으십시오.
```

```python
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]

plt.imshow(img)
plt.colorbar()
plt.show()

#여기서 꽤 많은 매개 변수를 가진 cv2.grabCut을 사용함
#먼저 입력 이미지, 마스크, 주 개체의 사각형, 배경 모델, 전경 모델, 반복 반복 실행 양 및 사용중인 모드를 선택합니다.

#여기에서 마스크가 변경되어 모든 0 및 2 픽셀이 배경으로 변환되며 1 및 3 픽셀이 이제 전경이됩니다. 
#여기에서 입력 이미지를 곱하면 최종 결과를 얻을 수 있습니다.
```
